```python
#1.

# Load the datasets
gapminder = pd.read_csv("gapminder_gapminder.csv")
gapminder_full = pd.read_csv("gapminder_full.csv")
# Display the first few rows of each dataset
gapminder_head = gapminder.head()
gapminder_full_head = gapminder_full.head()
gapminder_head, gapminder_full_head
```

```python
#to find out how many unique countries are there
num_countries_gapminder = gapminder['country'].nunique()
num_countries_gapminder_full = gapminder_full['country'].nunique()
num_countries_gapminder, num_countries_gapminder_full
```

```python
#2.
# Create pivot table that shows the average life expectancy for each continent and year.
pivot_table = gapminder_full.pivot_table(
    values='life_exp',
    index='continent',
    columns='year',
    aggfunc='mean'
)
pivot_table
```

```python
#3.
# Find countries with GDP per capita higher than the 75th percentile
gapminder_2007 = gapminder[gapminder['year'] == 2007]
gdp_per_capita_75th = gapminder['gdp_cap'].quantile(0.75)
high_gdp_countries = gapminder[gapminder['gdp_cap'] > gdp_per_capita_75th]['country']
high_gdp_countries.tolist()
```

```python
#4.
# Define the bin labels and Create the 'Life_Exp_Range' column
bin_labels = ['Low', 'Medium', 'High', 'Very High']
gapminder_full['Life_Exp_Range'] = pd.cut(gapminder_full['life_exp'], bins=4, labels=bin_labels)
gapminder_full
```

```python
#5.
# Sort the data by GDP per capita in descending order and select the top 5 countries
import matplotlib.pyplot as plt
#gapminder_2007 = gapminder_full[gapminder_full['year'] == 2007]
top_5_countries = gapminder.nlargest(5, 'gdp_cap')
plt.figure(figsize=(10,6))
plt.barh(top_5_countries['country'], top_5_countries['gdp_cap'], color='skyblue')
plt.xlabel('GDP per Capita')
plt.title('Top 5 Countries by GDP per Capita in 2007')
plt.gca().invert_yaxis()
plt.show()
```

```python
#6.
# Find country names that start with "I" and end with "a"
import re
pattern = r'^I.*a$'
countries_with_I_a = gapminder_full[gapminder_full['country'].str.contains(pattern, regex=True)]['country'].unique()
countries_with_I_a
```

```python
#7.
# Filter dataset for the year 2007
# Create a boxplot
import seaborn as sns
import matplotlib.pyplot as plt
#gapminder_2007 = gapminder[gapminder['year'] == 2007]
plt.figure(figsize=(10,6))
sns.boxplot(x='continent', y='gdp_cap', data=gapminder)
plt.title('GDP per Capita Distribution by Continent in 2007')
plt.xlabel('Continent')
plt.ylabel('GDP per Capita')
plt.show()
```

```python
#8.
# Filter dataset for countries with life expectancy over 80 in 2007
high_life_exp_countries = gapminder[gapminder['life_exp'] > 80][['country', 'continent']]
high_life_exp_countries
```

```python
#9.
# Convert 'year' column to datetime type and extract the decade
gapminder_full['year'] = pd.to_datetime(gapminder_full['year'], format='%Y')
gapminder_full['Decade'] = (gapminder_full['year'].dt.year // 10) * 10
gapminder_full[['year', 'Decade']].head()
```

```python
#10.
# Compute the correlation matrix
# Create a heatmap
correlation_matrix = gapminder_full[['gdp_cap', 'life_exp', 'population']].corr()
correlation_matrix
```

```python
plt.figure(figsize=(12,8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', vmin=-1, vmax=1)
plt.title('Correlation Matrix')
plt.show()
```

```python
#11.
# Calculate the global average life expectancy for each year
global_life_expectancy = gapminder_full.groupby('year')['life_exp'].mean().reset_index()
# Plot a line graph
plt.figure(figsize=(12,8))
plt.plot(global_life_expectancy['year'], global_life_expectancy['life_exp'], marker='o')
plt.title('Global Average Life Expectancy (1952-2007)')
plt.xlabel('Year')
plt.ylabel('Life Expectancy')
plt.grid(True)
plt.show()
```

```python
#12.
# Create a scatter plot
plt.figure(figsize=(12,8))
sns.scatterplot(x='gdp_cap', y='life_exp', data=gapminder)
plt.title('Life Expectancy vs GDP per Capita in 2007')
plt.xlabel('GDP per Capita')
plt.ylabel('Life Expectancy')
plt.grid(True)
plt.show()
```

```python
#13.
# Calculate average GDP per capita for each continent in 2007
average_gdp_continent = gapminder.groupby('continent')['gdp_cap'].mean().reset_index()
# Create a bar chart
plt.figure(figsize=(4,4))
sns.barplot(x='continent', y='gdp_cap', data=average_gdp_continent, palette='viridis')
plt.title('Average GDP per Capita by Continent in 2007')
plt.xlabel('Continent')
plt.ylabel('Average GDP per Capita')
plt.show()
```

```python
#14.
#1)

# Filter dataset for Afghanistan and Switzerland for 2007
afg_swi_2007 = gapminder[gapminder['country'].isin(['Afghanistan', 'Switzerland'])]

# Create bar charts for life expectancy and GDP per capita in 2007
fig, axes = plt.subplots(1, 2, figsize=(12,16))

# Life expectancy
sns.barplot(x='country', y='life_exp', data=afg_swi_2007, ax=axes[0], palette='Set2')
axes[0].set_title('Life Expectancy in 2007')
axes[0].set_xlabel('Country')
axes[0].set_ylabel('Life Expectancy')

# GDP per capita
sns.barplot(x='country', y='gdp_cap', data=afg_swi_2007, ax=axes[1], palette='Set2')
axes[1].set_title('GDP per Capita in 2007')
axes[1].set_xlabel('Country')
axes[1].set_ylabel('GDP per Capita')

plt.tight_layout()
plt.show()
```

```python
#2)

# Filter dataset for Afghanistan and Switzerland for all years
afg_swi_all_years = gapminder_full[gapminder_full['country'].isin(['Afghanistan', 'Switzerland'])]

# Create line graphs for life expectancy and GDP per capita trends
fig, axes = plt.subplots(2, 1, figsize=(12,16))

# Life expectancy trend
sns.lineplot(x='year', y='life_exp', hue='country', data=afg_swi_all_years, ax=axes[0])
axes[0].set_title('Life Expectancy Trend (1952-2007)')
axes[0].set_xlabel('Year')
axes[0].set_ylabel('Life Expectancy')

# GDP per capita trend
sns.lineplot(x='year', y='gdp_cap', hue='country', data=afg_swi_all_years, ax=axes[1])
axes[1].set_title('GDP per Capita Trend (1952-2007)')
axes[1].set_xlabel('Year')
axes[1].set_ylabel('GDP per Capita')

plt.tight_layout()
plt.show()
```
